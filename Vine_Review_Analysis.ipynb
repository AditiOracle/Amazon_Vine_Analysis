```python
import pandas as pd
import os
```

```python
#Read vine tabe data as datafram
vine_data="vine_data.csv"
df=pd.read_csv(vine_data)
df.head()

```

```python
#total_votes greater than or equql to 20 
total_votes_df=df.loc[(df["total_votes"])>=20,:]
total_votes_df
```

```python
#helpful_votes divided by total_votes is equal to or greater than 50%
helpful_votes_df=total_votes_df.loc[(total_votes_df["helpful_votes"])/(total_votes_df["total_votes"])*100>=50,:]
helpful_votes_df
```

```python
#Retrieve rows where vine program is paid(==Y)
vine_paid=helpful_votes_df.loc[(helpful_votes_df["vine"])=="Y",:]
vine_paid
```

```python
#Retrieve rows where vine program is unpaid(==N)
vine_unpaid=helpful_votes_df.loc[(helpful_votes_df["vine"])=="N",:]
vine_unpaid
```

```python
#Count paid vine program reviews where start rating is 5
vine_paid_review_count=vine_paid.loc[vine_paid["star_rating"]==5.0].count()["review_id"]
vine_paid_review_count
```

```python
#Count paid vine program reviews
vine_paid_total_reviews=vine_paid["review_id"].count()
vine_paid_total_reviews
```

```python
#Calculate percentage of paid 5-start rating vine program
vine_paid_five_star_percent=(vine_paid_review_count/vine_paid_total_reviews)*100
round(vine_paid_five_star_percent,2)
```

```python
#Count unpaid vine program reviews where start rating is 5
vine_unpaid_review_count=vine_unpaid.loc[vine_unpaid["star_rating"]==5.0].count()["review_id"]
vine_unpaid_review_count
```

```python
#Count unpaid vine program reviews
vine_unpaid_total_reviews=vine_unpaid["review_id"].count()
vine_unpaid_total_reviews
```

```python
#Calculate percentage of unpaid 5-start rating vine program
vine_unpaid_five_star_percent=(vine_unpaid_review_count/vine_unpaid_total_reviews)*100
round(vine_unpaid_five_star_percent,2)
```
